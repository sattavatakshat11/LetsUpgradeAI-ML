```python
# Null Hyptohesis : the mean distance from the home is 8.900
# Alternative hypothesis  : the mean distance from home is not 8.900

# 2 - Statistical Test:
#     z test :
#         Calculated value  = 0.07
#         table value is 2.086
#         so confidence interval is -2.086 - +2.086
#         calculated value falls within the given interval so null hypothesis is accepted!
        
        
```

```python
nd = ['Age', 'DistanceFromHome', 'MonthlyIncome',
       'NumCompaniesWorked',  'PercentSalaryHike', 'StandardHours',
        'TotalWorkingYears', 'TrainingTimesLastYear',
       'YearsAtCompany', 'YearsSinceLastPromotion', 'YearsWithCurrManager']
```

```python
def hypo(a):
    table_value_positive = 2.086
    table_value_negative = -2.086
    for i in a:
        
        print("Null Hypothesis: The average mean of the poulation parameter" , i ,"=",asamp[i].mean(),"\n")
        print("Alternative Hypothesis: The average mean of the poulation parameter" , i ,"is not equal to",asamp[i].mean(),"\n")
        
        print("The average value of dataset is",df[i].mean(),"\n")
        print("The average value of the sample is " , asamp[i].mean(),"\n")
        
        calculated_value = (int(asamp[i].mean()) - int(df[i].mean())) / (df[i].std() / np.sqrt(20))
        print("calculated value :" , calculated_value)
        print("Table Value :",table_value_negative , table_value_positive)
        
        if (calculated_value < table_value_positive) | (calculated_value > table_value_negative):
            print("###################################################")
            print("Null hypothesis of " , i, "is accepted", " \n")
            print("###################################################")
            print("\n")
        else:
            print("Alternative hypothesis of", i , "is accepted")
            print("\n")

        
```

```python
hypo(nd)
```
